```python
first_sorted = first_dose.sort_values('First Dose Administered').copy()
first_sorted['First Dose (Lakhs)'] = first_sorted['First Dose Administered'] / 100000

plt.figure(figsize=(12,6))

plt.bar(first_sorted['State'],
        first_sorted['First Dose (Lakhs)'])

plt.xticks(rotation=90)
plt.ylabel("First Dose Administered (in Lakhs)")
plt.title("State-wise First Dose Vaccination (in Lakhs)")
plt.tight_layout()
plt.show()
```

```python
second_sorted = second_dose.sort_values('Second Dose Administered').copy()
second_sorted['Second Dose (Lakhs)'] = second_sorted['Second Dose Administered'] / 100000

plt.figure(figsize=(12,6))

plt.bar(second_sorted['State'],
        second_sorted['Second Dose (Lakhs)'])

plt.xticks(rotation=90)
plt.ylabel("Second Dose Administered (in Lakhs)")
plt.title("State-wise Second Dose Vaccination (in Lakhs)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Remove India row
df_states = df[df['State'] != 'India']

# Get latest record per state
latest_data = df_states.sort_values('Updated On').groupby('State').tail(1)

# Calculate totals
total_male = latest_data['Male (Doses Administered)'].sum()
total_female = latest_data['Female (Doses Administered)'].sum()

# Convert to Lakhs
total_male_lakhs = total_male / 100000
total_female_lakhs = total_female / 100000

values = [total_male_lakhs, total_female_lakhs]
labels = ['Male (Lakhs)', 'Female (Lakhs)']

# Function to show actual numbers instead of %
def show_values(pct):
    total = sum(values)
    val = pct * total / 100
    return f'{val:.1f}'

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct=show_values,colors=['skyblue','pink'])

plt.title("Male vs Female Vaccination Distribution (in Lakhs)")
plt.show()
```
